```python
# Data handling imports

import yfinance as yf
import pandas as pd
import numpy as np
# import datetime as dt
from datetime import date
# import datetime
```

```python
tix = pd.read_table('./DJIA_comps.txt', header=None).to_numpy()
```

```python
tix = np.concatenate(tix, axis=0).tolist()
```

```python
import pickle

with open('tix_df.pkl', 'rb') as f:
    tix_df = pickle.load(f)    
```

```python
date = tix_df[0].iloc[0,0]
```

```python
stats = ['trendingScore', 'sentimentChange', 'volumeChange', 'percent', 'extendedHoursPercentChange']
```

```python
stats_df = pd.DataFrame(columns=stats, index=tix)
```

```python
stats_df
```

```python
for i in range(len(tix_df)):
    for stat in stats:
        stats_df.loc["{}".format(tix[i]), stat] = tix_df[i].loc[0, stat]
    
```

```python
stats_df
```

```python
import yfinance as yf
```

```python
tix_return = []
```

```python
for tick in tix:
    tick_data = yf.download(tick, date.today())
    tick_return = tick_data['Adj Close']/tick_data['Open'] - 1
    tix_return.append(tick_return.values[0]*100)
```

```python
tix_return_bin = []

for ret in tix_return:
    if ret > 0:
        tix_return_bin.append(1)
    else:
        tix_return_bin.append(-1)
```

```python
feats = stats_df.iloc[:,4].to_numpy()
feats = feats.reshape(-1, 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
from sklearn import tree
```

```python
acc = []
for _ in range(100):
    x_train, x_test, y_train, y_test = train_test_split(feats, tix_return_bin, test_size=0.30)

#     model = RandomForestClassifier(n_estimators=10, bootstrap=False,
#                                    max_features='sqrt', max_depth=10)
    
#     12, 10 for just pct chg during and after
    model = tree.DecisionTreeClassifier(max_depth=1)
#     model = tree.DecisionTreeRegressor(max_depth=1)
    
    model.fit(x_train, y_train)

    y_pred = model.predict(x_test).tolist()
#     y_test = y_test

    acc.append(accuracy_score(y_test, y_pred))
#     acc.append(model.score(x_test,y_test))
```
