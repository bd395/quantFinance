```python
for w in range(len(stats_df)):
    tick = stats_df.iloc[w].name.split("_")[0]
    date = stats_df.iloc[w].name.split("_")[1]
    date_dt = dt.datetime.strptime(date, '%Y-%m-%d')
    date_stamp = pd.Timestamp(date)
    weekday = date_stamp.dayofweek
    
    if weekday == 4:
        delta = timedelta(days=3)
    else:
        delta = timedelta(days=1)
    
    nextday = date_dt + delta
    
    tick_data = yf.download(tick, nextday, nextday+timedelta(days=1))
    
    tick_return = tick_data.iloc[0, 4]/tick_data.iloc[0,0] - 1
    
    tix_return.iloc[w] = tick_return
#     tix_return.append(tick_return)

# for tick in tix:
#     tick_data = yf.download(tick, date.today()-timedelta(days=1), date.today()+timedelta(days=1))
    
#     day1 = tick_data.loc[date.today()-timedelta(days=1)]
#     day2 = tick_data.loc[date.today()]
    
#     tick_return1 = day1.loc['Adj Close']/day1['Open'] - 1
#     tick_ret1 = tick_return1*100
#     tix_return.append(tick_ret1)
    
#     tick_return2 = day2.loc['Adj Close']/day2['Open'] - 1
#     tick_ret2 = tick_return2*100
#     tix_return.append(tick_ret2)

    
#     tix_return.append(tick_return.values[0]*100)
```

```python
acc = []
long_acc = []
long_rets = []

for _ in range(100):
    x_train, x_test, y_train, y_test = train_test_split(feats, labels, test_size=0.33)

#     model = RandomForestClassifier(n_estimators=100, bootstrap=False,
#                                    max_features='sqrt', max_depth=10)

#     model = MLPClassifier(solver='lbfgs', alpha=1e-5,
#                               hidden_layer_sizes=(30, 30),#(25,25,5), #10, 10, 5
#                               activation='relu',
#                               max_iter=2000)    
    
#     12, 10 for just pct chg during and after
#     model = tree.DecisionTreeClassifier(max_depth=7)
#     model = LogisticRegression(penalty="l2")
#     model = tree.DecisionTreeRegressor(max_depth=1)
#     model = SVC(kernel='rbf')
    model = GaussianNB(priors=[0.5,0.5])
    
    model.fit(x_train, y_train)

    y_pred = model.predict(x_test)
#     y_test = y_test

    acc.append(accuracy_score(y_test, y_pred))
#     acc.append(model.score(x_test,y_test))

    y_test_list = y_test.tolist()
    did_long = 0
    should_long = 0
    long_idx = []
    
    for i in range(len(y_pred)):
        if y_pred[i] == 1:
            did_long += 1
            long_idx.append(y_test.index[i])
            if y_test_list[i] == 1:
                should_long +=1
    
    long_ret = []
    
    for idx in long_idx:
        long_ret.append(tix_return.iloc[idx])
    
#     print(long_ret)
#     print(" ")
    
    long_rets.append(np.mean(long_ret))    
    
#     print(y_test_list)
#     print(should_long)
#     print(y_pred)
#     print(did_long)
    
    long_acc.append(should_long/did_long)
```

```python
print('Mean: ', np.mean(acc))
print("Min: ", min(acc))
print("Max: ", max(acc))
# print("Mode: ", mode(acc))
print("Long Acc: ", np.mean(long_acc))
print("Average long return: ", np.mean(long_rets))
```
