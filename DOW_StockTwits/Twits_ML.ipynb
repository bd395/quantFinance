```python
stats = ['trendingScore', 'sentimentChange', 'volumeChange', 'percent', 'extendedHoursPercentChange']
```

```python
stats_df = pd.DataFrame(columns=stats)
```

```python
stats_df
```

```python
for i in range(len(tix_df)):
    for stat in stats:
        stats_df.loc["{}_1".format(tix[i]), stat] = tix_df[i].loc[0, stat]
        stats_df.loc["{}_2".format(tix[i]), stat] = tix_df[i].loc[1, stat]
    
```

```python
stats_df
```

```python
import yfinance as yf
```

```python
tix_return = []
```

```python
for tick in tix:
    tick_data = yf.download(tick, date.today()-timedelta(days=1), date.today()+timedelta(days=1))
    day1 = tick_data.loc[date.today()-timedelta(days=1)]
    day2 = tick_data.loc[date.today()]
    
    tick_return1 = day1.loc['Adj Close']/day1['Open'] - 1
    tick_ret1 = tick_return1*100
    tix_return.append(tick_ret1)
    
    tick_return2 = day2.loc['Adj Close']/day2['Open'] - 1
    tick_ret2 = tick_return2*100
    tix_return.append(tick_ret2)

    
#     tix_return.append(tick_return.values[0]*100)
```

```python
len(tix_return)
```

```python
tix_return_bin = []

for ret in tix_return:
    if ret > 0:
        tix_return_bin.append(1)
    else:
        tix_return_bin.append(-1)
    
```

```python
labels = pd.Series(tix_return_bin)

feats = stats_df.iloc[:,[0,3,4]]#.to_numpy()
# feats = feats.reshape(-1, 1)


# GaussianNB > SVC slightly better > than Logistic Regression (penalty=L2)
# GNB - (0),4 = 0.76
# GNB - 3, 4 = 0.72
# SVC - 4  = 0.73
# LR - 3 and 4 = 0.75
# LR - 4 = 0.7
# LR - 0 and 4 = 0.7
# LR - 0, 3 and = 0.66
```

```python
acc = []
long_acc = []

for _ in range(100):
    x_train, x_test, y_train, y_test = train_test_split(feats, labels, test_size=0.33)

#     model = RandomForestClassifier(n_estimators=20, bootstrap=False,
#                                    max_features='sqrt', max_depth=5)
    
#     12, 10 for just pct chg during and after
#     model = tree.DecisionTreeClassifier(max_depth=2)
#     model = LogisticRegression(penalty="l2")
#     model = tree.DecisionTreeRegressor(max_depth=1)
#     model = SVC(kernel='rbf')
    model = GaussianNB()#priors=[0.2,0.8])
    
    model.fit(x_train, y_train)

    y_pred = model.predict(x_test).tolist()
#     y_test = y_test

    acc.append(accuracy_score(y_test, y_pred))
#     acc.append(model.score(x_test,y_test))

    y_test_list = y_test.tolist()
    did_long = 0
    should_long = 0
    for i in range(len(y_pred)):
        if y_pred[i] == 1:
            did_long += 1
            if y_test_list[i] == 1:
                should_long +=1

    
#     print(y_test_list)
#     print(should_long)
#     print(y_pred)
#     print(did_long)
    
    long_acc.append(should_long/did_long)
```

```python
print('Mean: ', np.mean(acc))
print("Min: ", min(acc))
print("Max: ", max(acc))
#print("Mode: ", mode(acc))
print("Long Acc: ", np.mean(long_acc))
```
